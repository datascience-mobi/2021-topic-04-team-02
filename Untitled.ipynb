```python
train_labels, train_values = load_the_pickle('data/train_points.p')
c = train_values - np.mean(train_values, axis = 0)
m1 = np.cov(train_values, rowvar = 0)
m2 = np.cov(c, rowvar = 0)
```

```python
def covvey(mat):
    x = len(mat[0])
    c = (1/(x-1))
    covvmat = np.zeros((x, x))
    for i in range(0, x):
        xi_xmean = (mat[:,i] - np.mean(mat[:,i]))
        for j in range (0, x):
            covvmat[i,j] = c * np.dot( xi_xmean , (mat[:,j] - np.mean(mat[:,j])))
        
    return covvmat
```

```python
print(np.array_equal(covvey(c), m2))
```
